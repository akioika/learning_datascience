```python
# 必要なライブラリを読み込む
import pandas as pd
```

```python
# Excel を Windows エンコードで読み込む
# シートごとに読み込む必要があるが、シート名を指定するとエンコードの問題が発生する可能背があるので
# シート番号を指定する
d_name = "/root/work/excel_data_processing/"
f_name = d_name + "sample_data.xlsx"
data_sheet = pd.read_excel(f_name, sheet_name=0, encoding="cp932")
master_sheet = pd.read_excel(f_name, sheet_name=1, encoding="cp932")

# SJIS な Excel を適切に処理できないので、一旦何もしないで BOM なし UTF-8 な CSV にする
data_sheet.to_csv(d_name + 'data.csv', index=False, encoding='utf-8')
master_sheet.to_csv(d_name + 'master.csv', index=False, encoding='utf-8')
```

```python
# 読みなおし
data_df = pd.read_csv(d_name + 'data.csv')
master_df = pd.read_csv(d_name + 'master.csv')

# マスタから値段カラムを data.csv に追加
data_df = pd.merge(data_df, master_df, how='inner', left_on='食べたもの', right_on='メニュー')
print(data_df)
```

```python
# のりしろ部分のカラムを削除
data_df = data_df.drop('メニュー', axis=1)
print(data_df)
```

```python
# sort
data_df = data_df.sort_values(by=['食べたもの','食べた日'], ascending=[True,False])
# 連番をふる
data_df['ユニークカウント'] = data_df.groupby('食べたもの').cumcount()+1
# カラム順変更
data_df = data_df.loc[:,['ユニークカウント','食べたもの','食べた日','食べた場所','値段']]
print(data_df)
```

```python
# BOM あり UTF-8 な CSV に出力
data_df.to_csv(d_name + 'result.csv', index=False, encoding='utf_8_sig')
# ユニークカウントが 1 の者のみ BOM あり UTF-8 な CSV に出力
data_df[data_df['ユニークカウント'].isin([1])].to_csv(d_name + 'result_U1.csv', index=False, encoding='utf_8_sig')
```
